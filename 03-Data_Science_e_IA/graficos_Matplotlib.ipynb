📊 Tipos de Gráficos en Python: Conceptos, Ejemplos y Cuándo Usarlos

En esta sección, exploraremos diferentes tipos de gráficos utilizados en Python para visualizar datos, explicando su concepto, proporcionando ejemplos prácticos en código y detallando cuándo se deben usar.

🔹 1. Gráfico de Líneas (plot)

🔍 Concepto:

Los gráficos de líneas son ideales para representar datos continuos a lo largo del tiempo, mostrando tendencias y patrones.

📅 Cuándo Usarlo:

Para analizar series temporales.

Para observar la tendencia de una variable a lo largo del tiempo.

Comparar múltiples tendencias en un mismo gráfico.

🔄 Ejemplo:

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y, label='Seno', color='blue')
plt.title("Gráfico de Líneas")
plt.xlabel("Tiempo")
plt.ylabel("Valor")
plt.legend()
plt.show()
```

🔹 2. Gráfico de Barras (bar)

🔍 Concepto:

Se usa para comparar diferentes categorías o mostrar la frecuencia de ocurrencias de diferentes elementos.

📅 Cuándo Usarlo:

Comparar cantidades entre diferentes categorías.

Representar datos categóricos.

Mostrar la frecuencia de valores discretos.

🔄 Ejemplo:

```python
categorias = ["A", "B", "C", "D"]
valores = [5, 7, 3, 8]

plt.bar(categorias, valores, color='purple')
plt.title("Gráfico de Barras")
plt.xlabel("Categorías")
plt.ylabel("Valores")
plt.show()
```

🔹 3. Histograma (hist)

🔍 Concepto:

Muestra la distribución de una variable numérica dividiéndola en intervalos (bins).

📅 Cuándo Usarlo:

Analizar la distribución de una variable.

Identificar la presencia de valores atípicos.

Determinar si una variable sigue una distribución normal.

🔄 Ejemplo:

```python
datos = np.random.randn(1000)

plt.hist(datos, bins=30, color='green', alpha=0.7)
plt.title("Histograma de Distribución")
plt.xlabel("Valores")
plt.ylabel("Frecuencia")
plt.show()
```

🔹 4. Gráfico de Dispersión (scatter)

🔍 Concepto:

Muestra la relación entre dos variables numéricas con puntos dispersos.

📅 Cuándo Usarlo:

Visualizar correlaciones entre variables.

Identificar patrones o tendencias en datos numéricos.

Detectar valores atípicos.

🔄 Ejemplo:

```python
x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x, y, color='red', alpha=0.6)
plt.title("Gráfico de Dispersión")
plt.xlabel("Variable X")
plt.ylabel("Variable Y")
plt.show()
```

🔹 5. Gráfico de Torta (pie)

🔍 Concepto:

Se usa para representar proporciones o porcentajes en un conjunto de datos.

📅 Cuándo Usarlo:

Representar porcentajes de una variable categórica.

Comparar la participación de cada categoría en el total.

🔄 Ejemplo:

```python
etiquetas = ['A', 'B', 'C', 'D']
valores = [30, 25, 35, 10]

plt.pie(valores, labels=etiquetas, autopct='%1.1f%%', colors=['gold', 'blue', 'red', 'green'])
plt.title("Gráfico de Torta")
plt.show()
```

2️⃣ ⭐️ Gráficos de Distribución en Python: Conceptos, Ejemplos y Cuándo Usarlos

En esta sección, exploraremos los diferentes tipos de gráficos de distribución utilizados en Python para visualizar la distribución de datos, identificación de valores atípicos y patrones en un conjunto de datos.

🔹 1. Boxplot (Diagrama de Cajas) (boxplot)

🔍 Concepto:

Un boxplot muestra la distribución de una variable mediante un resumen estadístico (mínimo, primer cuartil, mediana, tercer cuartil y máximo). También permite identificar valores atípicos.

📅 Cuándo Usarlo:

Para visualizar la distribución y dispersión de una variable numérica.

Para identificar valores atípicos (outliers).

Para comparar distribuciones entre diferentes grupos.

🔄 Ejemplo:

```python
import matplotlib.pyplot as plt
import numpy as np

datos = np.random.randn(100)

plt.boxplot(datos)
plt.title("Boxplot - Diagrama de Cajas")
plt.ylabel("Valores")
plt.show()
```

🔹 2. Violin Plot (violinplot)

🔍 Concepto:

Un violin plot combina la información del boxplot con una distribución de densidad, proporcionando una visión más detallada de la distribución de los datos.

📅 Cuándo Usarlo:

Cuando se necesita analizar la forma y simetría de la distribución.

Para comparar distribuciones entre varios grupos.

Para ver si los datos tienen distribución bimodal o multimodal.

🔄 Ejemplo:

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

datos = np.random.randn(100)
sns.violinplot(y=datos)
plt.title("Violin Plot")
plt.show()
```

🔹 3. Stem Plot (stem)

🔍 Concepto:

Un stem plot representa datos discretos mediante líneas verticales desde el eje X, facilitando la visualización de distribuciones de valores en ciertos puntos.

📅 Cuándo Usarlo:

Cuando se quiere visualizar series temporales con valores discretos.

Para destacar valores individuales en una serie de datos.

Para analizar patrones en datos con puntos específicos.

🔄 Ejemplo:

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 10)
y = np.sin(x)

plt.stem(x, y)
plt.title("Stem Plot")
plt.xlabel("Tiempo")
plt.ylabel("Valores")
plt.show()
```

🔹 4. Gráfico de Pirámide Poblacional

🔍 Concepto:

Este tipo de gráfico se utiliza para comparar distribuciones en diferentes categorías, generalmente en estudios demográficos.

📅 Cuándo Usarlo:

Para analizar la distribución de poblaciones por edad y género.

Para comparar dos grupos en una misma categoría.

En estudios de mercado para analizar segmentos de clientes.

🔄 Ejemplo:

```python
import matplotlib.pyplot as plt

categorias = ['0-18', '19-35', '36-50', '51+']
mujeres = [30, 40, 35, 20]
hombres = [35, 45, 30, 25]

plt.barh(categorias, mujeres, color='pink', label='Mujeres')
plt.barh(categorias, [-h for h in hombres], color='blue', label='Hombres')
plt.axvline(0, color='black', linewidth=0.8)
plt.title("Pirámide Poblacional")
plt.legend()
plt.show()
```

3️⃣ 🌟 Gráficos Avanzados en Python: Conceptos, Ejemplos y Cuándo Usarlos

En esta sección, exploraremos los diferentes tipos de gráficos avanzados utilizados en Python para visualizar datos de manera más detallada y obtener insights profundos sobre patrones y relaciones en los datos.

🔹 1. Heatmap (Mapa de Calor) (imshow)

🔍 Concepto:

Un heatmap es una representación en color de datos organizados en una matriz, donde los colores indican la magnitud de los valores.

📅 Cuándo Usarlo:

Para analizar correlaciones entre variables en un dataset.

Para visualizar distribuciones de datos en una matriz.

Para analizar datos en cuadrículas (ejemplo: modelos geoespaciales o de redes neuronales).

🔄 Ejemplo:

```python
import matplotlib.pyplot as plt
import numpy as np

matriz = np.random.rand(10, 10)
plt.imshow(matriz, cmap='viridis', interpolation='nearest')
plt.colorbar()
plt.title("Heatmap - Mapa de Calor")
plt.show()
```

🔹 2. Gráfico de Contorno (contour)

🔍 Concepto:

El gráfico de contorno muestra las líneas de nivel de una función 3D en un espacio 2D.

📅 Cuándo Usarlo:

Para representar superficies en 3D en un plano 2D.

Para visualizar distribuciones de densidad en modelos de datos continuos.

En aplicaciones geoespaciales y modelado científico.

🔄 Ejemplo:

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

plt.contour(X, Y, Z, levels=10, cmap='plasma')
plt.title("Gráfico de Contorno")
plt.colorbar()
plt.show()
```

🔹 3. Gráfico de Contorno Relleno (contourf)

🔍 Concepto:

Similar al gráfico de contorno, pero con áreas rellenas de color para representar los niveles de magnitud.

📅 Cuándo Usarlo:

Para resaltar mejor la diferencia entre regiones de valores en un conjunto de datos continuos.

Para visualizar superficies de energía en modelado científico.

🔄 Ejemplo:

```python
plt.contourf(X, Y, Z, levels=20, cmap='coolwarm')
plt.title("Gráfico de Contorno Relleno")
plt.colorbar()
plt.show()
```

🔹 4. Quiver Plot (quiver)

🔍 Concepto:

El quiver plot representa un campo vectorial en un espacio bidimensional mediante flechas.

📅 Cuándo Usarlo:

Para representar direcciones y magnitudes en un campo vectorial (ejemplo: vientos en meteorología).

En modelado de flujos de fluidos y simulaciones físicas.

🔄 Ejemplo:

```python
X, Y = np.meshgrid(np.arange(0, 10, 1), np.arange(0, 10, 1))
U = np.sin(X)
V = np.cos(Y)

plt.quiver(X, Y, U, V)
plt.title("Quiver Plot - Campo Vectorial")
plt.show()
```

🔹 5. Streamplot (streamplot)

🔍 Concepto:

El streamplot muestra el flujo de partículas en un campo vectorial continuo.

📅 Cuándo Usarlo:

Para analizar flujos de fluidos en dinámica de fluidos computacional.

Para representar direcciones y trayectorias en simulaciones físicas.

🔄 Ejemplo:

```python
Y, X = np.mgrid[0:10:100j, 0:10:100j]
U = np.cos(X)
V = np.sin(Y)

plt.streamplot(X, Y, U, V, color=U, linewidth=2, cmap='viridis')
plt.title("Streamplot - Flujo de Partículas")
plt.show()
```

🔹 6. Error Bars (errorbar)

🔍 Concepto:

El error bar plot agrega barras de error a los puntos de datos para representar la incertidumbre de las mediciones.

📅 Cuándo Usarlo:

En experimentos científicos donde hay incertidumbre en las mediciones.

Para visualizar intervalos de confianza en un conjunto de datos.

🔄 Ejemplo:

```python
x = np.linspace(0, 10, 10)
y = np.sin(x)
errores = np.random.rand(10) * 0.1

plt.errorbar(x, y, yerr=errores, fmt='o', capsize=5)
plt.title("Error Bars - Barras de Error")
plt.show()
```

✨ Gráficos 3D en Python: Conceptos, Ejemplos y Cuándo Usarlos

Los gráficos 3D nos permiten visualizar datos tridimensionales de manera clara y efectiva. Son útiles para representar funciones matemáticas, datos volumétricos, visualizaciones geoespaciales y mucho más.

💡 1. Gráfico de Superficie (Surface Plot)

💡 Concepto:

Un gráfico de superficie muestra la variación de una función en un espacio tridimensional. Se usa para representar relaciones continuas entre tres variables.

📅 Cuándo Usarlo:

Para visualizar funciones matemáticas en 3D.

En modelado de terrenos y simulaciones científicas.

Para representar datos que dependen de dos variables independientes.

🔄 Ejemplo:

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
plt.title("Gráfico de Superficie")
plt.show()
```

💡 2. Gráfico de Dispersión 3D (Scatter 3D)

💡 Concepto:

Un scatter plot 3D representa puntos de datos en un espacio tridimensional.

📅 Cuándo Usarlo:

Para representar nubes de puntos en análisis de datos.

En simulaciones de física y modelos de machine learning.

Para detectar agrupaciones en datos multivariados.

🔄 Ejemplo:

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = np.random.rand(100)
y = np.random.rand(100)
z = np.random.rand(100)

ax.scatter(x, y, z, c=z, cmap='coolwarm')
plt.title("Gráfico de Dispersión 3D")
plt.show()
```

💡 3. Gráfico de Líneas 3D

💡 Concepto:

Un gráfico de líneas en 3D se usa para representar trayectorias en un espacio tridimensional.

📅 Cuándo Usarlo:

Para visualizar movimientos en un espacio tridimensional.

En simulaciones de trayectorias y caminos.

Para representar datos secuenciales en 3D.

🔄 Ejemplo:

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

t = np.linspace(0, 10, 100)
x = np.sin(t)
y = np.cos(t)
z = t

ax.plot(x, y, z, label="Trayectoria 3D")
ax.legend()
plt.title("Gráfico de Líneas 3D")
plt.show()
```

💡 4. Wireframe Plot

💡 Concepto:

El wireframe plot es similar a un gráfico de superficie, pero usa líneas en lugar de superficies rellenas.

📅 Cuándo Usarlo:

Para representar estructuras tridimensionales con menos detalles.

En modelos científicos y arquitectónicos.

Cuando se necesita una visualización ligera sin sombreado.

🔄 Ejemplo:

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_wireframe(X, Y, Z, color='black')
plt.title("Wireframe Plot")
plt.show()
```

💡 5. Gráfico de Barras 3D

💡 Concepto:

El gráfico de barras 3D permite representar datos categóricos en tres dimensiones.

📅 Cuándo Usarlo:

Para representar comparaciones entre diferentes categorías en 3D.

En análisis de datos de negocios y visualización de KPIs.

Para visualizar distribuciones de frecuencias en varios ejes.

🔄 Ejemplo:

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = np.array([1, 2, 3, 4, 5])
y = np.array([1, 2, 3, 4, 5])
z = np.zeros(5)
dx = dy = np.ones(5)
dz = np.random.randint(1, 10, 5)

ax.bar3d(x, y, z, dx, dy, dz, color='blue')
plt.title("Gráfico de Barras 3D")
plt.show()
```

📊 Gráficos Especiales en Python

Los gráficos especiales en Python nos permiten visualizar datos en formatos menos convencionales, pero extremadamente útiles para ciertos análisis. A continuación, exploraremos algunos de los más importantes.

🔹 1. Gráfico de Radar (Radar Chart)

💡 Concepto:

El gráfico de radar es útil para comparar múltiples variables en forma de polígonos concéntricos.

📅 Cuándo Usarlo:

Para comparar características de diferentes categorías (ejemplo: rendimiento de atletas).

Para visualizar métricas de evaluación de modelos.

En análisis de encuestas y estudios de mercado.

🔄 Ejemplo:

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos
etiquetas = ['Precisión', 'Velocidad', 'Robustez', 'Escalabilidad', 'Eficiencia']
valores = [80, 70, 90, 85, 75]
valores += valores[:1]  # Cerrar el gráfico

# Ángulos
angulos = np.linspace(0, 2 * np.pi, len(etiquetas), endpoint=False).tolist()
angulos += angulos[:1]

# Gráfico
fig, ax = plt.subplots(figsize=(6,6), subplot_kw=dict(polar=True))
ax.fill(angulos, valores, color='blue', alpha=0.3)
ax.plot(angulos, valores, color='blue', linewidth=2)
ax.set_xticks(angulos[:-1])
ax.set_xticklabels(etiquetas)

plt.title("Gráfico de Radar")
plt.show()
```

🔹 2. Gráfico Ternario (Ternary Plot)

💡 Concepto:

El gráfico ternario representa tres variables que suman un total constante (ejemplo: proporción de tres componentes en una mezcla química).

📅 Cuándo Usarlo:

Para analizar proporciones de tres elementos que suman un total (100%).

En análisis de materiales, marketing y distribución de presupuestos.

Para evaluar estrategias de negocio que involucran tres factores.

🔄 Ejemplo:

📌 Nota: Se necesita instalar ternary con:

pip install python-ternary

```python
import matplotlib.pyplot as plt
import ternary

# Definir la escala del triángulo (suma de cada punto debe ser igual a 100)
scale = 100

# Crear la figura y el subplot para el gráfico ternario
fig, tax = ternary.figure(scale=scale)

# Dibujar los límites del triángulo
tax.boundary(linewidth=2.0)
tax.gridlines(color="gray", multiple=10)

# Definir los datos (cada conjunto de valores debe sumar 100)
datos = [(20, 50, 30), (40, 30, 30), (10, 70, 20)]

# Dibujar los puntos en el gráfico ternario
tax.scatter(datos, marker='o', color='red', label="Ejemplo de Datos")

# Agregar título y leyenda
tax.set_title("Gráfico Ternario")
tax.legend()

# Mostrar el gráfico
plt.show()


```

🔹 3. Histograma Circular (Rose Plot o Wind Rose)

💡 Concepto:

El histograma circular representa la frecuencia de datos distribuidos en diferentes direcciones o ángulos.

📅 Cuándo Usarlo:

Para analizar direcciones de viento en meteorología.

En estudios de comportamiento de tráfico y navegación.

Para visualizar patrones de datos angulares.

🔄 Ejemplo:

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos simulados
n = 100
angulos = np.random.uniform(0, 2*np.pi, n)
valores = np.random.rand(n)

# Gráfico
fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})
ax.hist(angulos, bins=12, weights=valores, color='green', alpha=0.7)
plt.title("Histograma Circular")
plt.show()
```

🔹 4. Stacked Area Chart (Gráfico de Área Apilada)

💡 Concepto:

El gráfico de área apilada muestra la evolución de varias categorías a lo largo del tiempo.

📅 Cuándo Usarlo:

Para representar el crecimiento de diferentes grupos en un conjunto de datos.

En análisis financieros y de ventas acumuladas.

Para mostrar cambios en proporciones de mercado.

🔄 Ejemplo:

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos simulados
x = np.arange(1, 11)
y1 = np.random.randint(5, 15, size=10)
y2 = np.random.randint(5, 15, size=10)
y3 = np.random.randint(5, 15, size=10)

# Apilamiento
plt.stackplot(x, y1, y2, y3, labels=['Categoría 1', 'Categoría 2', 'Categoría 3'], alpha=0.8)
plt.legend()
plt.title("Gráfico de Área Apilada")
plt.show()
```
