1锔 Histograma (histplot)

2锔 KDE Plot (kdeplot)
 Descripci贸n Expandida

El KDE Plot (Kernel Density Estimation) genera una curva que representa la densidad de los datos, suavizando la distribuci贸n y evitando la rigidez de los bins de un histograma.
 Cu谩ndo Usarlo

    Cuando se desea una representaci贸n suave y continua de la distribuci贸n.
    Para comparar m煤ltiples distribuciones en una sola gr谩fica.
    En casos donde los histogramas pueden parecer fragmentados.

锔 Excepciones

    No funciona bien con datos multimodales, ya que puede suavizar demasiado los picos.
    No es 煤til para datos categ贸ricos.
    Puede ser poco informativo si hay pocos datos.

 C贸digo Base

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

# Generar datos simulados
np.random.seed(42)
datos = np.random.normal(loc=50, scale=15, size=500)

# Crear KDE Plot con la sintaxis actualizada
sns.kdeplot(data=datos, fill=True, color="purple")  # Cambio de shade=True a fill=True
plt.title("KDE Plot de los Datos")
plt.xlabel("Valor")
plt.ylabel("Densidad")
plt.show()
```

 Explicaci贸n del C贸digo

    shade=True: Rellena el 谩rea bajo la curva.
    color="purple": Define el color de la l铆nea.

El ECDF (Empirical Cumulative Distribution Function) muestra la distribuci贸n acumulativa de los datos. Cada punto representa la fracci贸n de datos menores o iguales a un valor espec铆fico.
