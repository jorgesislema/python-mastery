📌 Gráficos de Distribución en Seaborn

Los gráficos de distribución permiten analizar cómo se distribuyen los valores de una variable numérica en un conjunto de datos. Son clave en el Análisis Exploratorio de Datos (EDA) y ayudan a visualizar patrones, sesgos y valores atípicos.
1️⃣ Histograma (histplot)
📝 Descripción Expandida

El histograma agrupa los valores en intervalos llamados bins y cuenta cuántos valores caen en cada intervalo. Es útil para visualizar la distribución de frecuencia.
📌 Cuándo Usarlo

    Para entender la forma general de la distribución de los datos.
    Cuando se busca identificar sesgos o valores extremos.
    Para comparar la frecuencia de valores en diferentes rangos.

⚠️ Excepciones

    No se recomienda con datos categóricos.
    Si el número de observaciones es muy bajo, puede ser poco representativo.
    Elegir un número incorrecto de bins puede ocultar patrones.

💻 Código Base

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

# Generar datos simulados con distribución normal
np.random.seed(42)
datos = np.random.normal(loc=50, scale=15, size=500)

# Crear histograma con KDE
sns.histplot(data=datos, bins=30, kde=True, color='blue')
plt.title("Histograma con KDE")
plt.xlabel("Valor")
plt.ylabel("Frecuencia")
plt.show()
```

🧐 Explicación del Código

    np.random.normal(loc=50, scale=15, size=500): Genera 500 valores con media 50 y desviación estándar 15.
    sns.histplot(data=datos, bins=30, kde=True, color='blue'):
        bins=30: Divide los datos en 30 grupos.
        kde=True: Agrega una estimación de densidad para suavizar la distribución.
        color='blue': Define el color del gráfico.

2️⃣ KDE Plot (kdeplot)
📝 Descripción Expandida

El KDE Plot (Kernel Density Estimation) genera una curva que representa la densidad de los datos, suavizando la distribución y evitando la rigidez de los bins de un histograma.
📌 Cuándo Usarlo

    Cuando se desea una representación suave y continua de la distribución.
    Para comparar múltiples distribuciones en una sola gráfica.
    En casos donde los histogramas pueden parecer fragmentados.

⚠️ Excepciones

    No funciona bien con datos multimodales, ya que puede suavizar demasiado los picos.
    No es útil para datos categóricos.
    Puede ser poco informativo si hay pocos datos.

💻 Código Base

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

# Generar datos simulados
np.random.seed(42)
datos = np.random.normal(loc=50, scale=15, size=500)

# Crear KDE Plot con la sintaxis actualizada
sns.kdeplot(data=datos, fill=True, color="purple")  # Cambio de shade=True a fill=True
plt.title("KDE Plot de los Datos")
plt.xlabel("Valor")
plt.ylabel("Densidad")
plt.show()
```

🧐 Explicación del Código

    shade=True: Rellena el área bajo la curva.
    color="purple": Define el color de la línea.

3️⃣ Boxplot (boxplot)
📝 Descripción Expandida

El boxplot o gráfico de caja y bigotes muestra:

    La mediana de los datos (línea en el centro de la caja).
    El rango intercuartil (caja).
    Valores atípicos (puntos fuera de los bigotes).

📌 Cuándo Usarlo

    Para detectar valores atípicos en un conjunto de datos.
    Cuando se necesita visualizar la mediana, cuartiles y dispersión.
    Comparar distribuciones entre diferentes categorías.

⚠️ Excepciones

    No proporciona información sobre la distribución exacta de los datos.
    Puede ser difícil de interpretar sin conocimiento estadístico.

💻 Código Base

```python
sns.boxplot(x=datos, color="orange")
plt.title("Boxplot de los Datos")
plt.xlabel("Valor")
plt.show()
```

🧐 Explicación del Código

    sns.boxplot(x=datos, color="orange"): Crea un boxplot de los datos con color naranja.

4️⃣ Violin Plot (violinplot)
📝 Descripción Expandida

El violin plot combina un boxplot con un KDE plot, proporcionando información sobre la distribución y densidad de los datos.
📌 Cuándo Usarlo

    Cuando se quiere visualizar tanto la densidad de los datos como la estadística descriptiva.
    Para comparar distribuciones entre grupos.

⚠️ Excepciones

    Puede ser difícil de interpretar para quienes no están familiarizados con KDE y boxplots.
    No es recomendable cuando se tienen pocos datos.

💻 Código Base

```python
sns.violinplot(x=datos, color="teal")
plt.title("Violin Plot de los Datos")
plt.xlabel("Valor")
plt.show()
```

🧐 Explicación del Código

    sns.violinplot(x=datos): Crea un violin plot de los datos.

5️⃣ ECDF Plot (ecdfplot)
📝 Descripción Expandida

El ECDF (Empirical Cumulative Distribution Function) muestra la distribución acumulativa de los datos. Cada punto representa la fracción de datos menores o iguales a un valor específico.
📌 Cuándo Usarlo

    Para entender la probabilidad acumulada de un conjunto de datos.
    Cuando se necesita analizar percentiles y valores extremos.
    Comparar la distribución entre múltiples conjuntos de datos.

⚠️ Excepciones

    No es un gráfico intuitivo para usuarios sin experiencia en estadística.
    No se recomienda para datos categóricos.

💻 Código Base

```python
sns.ecdfplot(data=datos, color="green")
plt.title("ECDF Plot de los Datos")
plt.xlabel("Valor")
plt.ylabel("Probabilidad Acumulada")
plt.show()
```

🧐 Explicación del Código

    sns.ecdfplot(data=datos, color="green"): Crea un gráfico ECDF con los datos en color verde.

🎯 Conclusiones Finales

✅ Para visualizar la frecuencia de valores: Usa un Histograma.
✅ Si quieres una distribución más suave: Usa un KDE Plot.
✅ Para detectar valores atípicos: Usa un Boxplot.
✅ Si deseas ver la densidad y la estadística descriptiva: Usa un Violin Plot.
✅ Para entender la acumulación de los datos: Usa un ECDF Plot.

📌 2️⃣ Gráficos de Relación en Seaborn

Los gráficos de relación permiten visualizar la conexión entre dos o más variables en un conjunto de datos. Son especialmente útiles en análisis exploratorio para identificar correlaciones y patrones en los datos.
1️⃣ Scatter Plot (Gráfico de Dispersión)
📌 Descripción expandida

Un scatter plot es un gráfico de dispersión que muestra la relación entre dos variables numéricas. Cada punto representa una observación en el conjunto de datos.
📌 Cuándo Usarlo

✅ Para visualizar relaciones entre dos variables continuas.
✅ Para detectar patrones, tendencias y correlaciones.
✅ Para identificar valores atípicos (outliers).
📌 Excepciones

❌ No es útil para datos categóricos.
❌ Si hay demasiados puntos, puede generar sobrecarga visual.
❌ Puede no ser efectivo si los datos tienen mucha dispersión y no muestran una tendencia clara.
📌 Código Base

```python
sns.scatterplot(x=variable_x, y=variable_y, data=datos)
```

📌 Explicación del Código:

    x=variable_x: Especifica la variable en el eje X.
    y=variable_y: Especifica la variable en el eje Y.
    data=datos: Indica el conjunto de datos.
📌 Ejemplo

```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Generar datos de ejemplo
np.random.seed(42)
datos = pd.DataFrame({
    "ingresos": np.random.rand(100) * 100000,  # Ingresos anuales en dólares
    "gastos": np.random.rand(100) * 80000      # Gastos anuales en dólares
})

# Crear scatter plot
sns.scatterplot(x="ingresos", y="gastos", data=datos, color="blue")
plt.title("Relación entre Ingresos y Gastos")
plt.xlabel("Ingresos ($)")
plt.ylabel("Gastos ($)")
plt.show()
```

2️⃣ Line Plot (Gráfico de Líneas)
📌 Descripción expandida

Un gráfico de líneas conecta puntos de datos en una secuencia, lo que lo hace ideal para visualizar tendencias a lo largo del tiempo.
📌 Cuándo Usarlo

✅ Para analizar tendencias en series temporales.
✅ Para ver el cambio de una variable en función de otra.
✅ Para comparar múltiples series de datos.
📌 Excepciones

❌ No es útil para datos sin una secuencia lógica (como categorías).
❌ Puede ser engañoso si hay muchas fluctuaciones sin una tendencia clara.
📌 Código Base

```python
sns.lineplot(x=variable_x, y=variable_y, data=datos)
```

📌 Explicación del Código:

    x=variable_x: Variable del eje X (normalmente tiempo).
    y=variable_y: Variable del eje Y (valor medido en cada instante).
    data=datos: Indica el conjunto de datos.

📌 Ejemplo

```python
# Generar datos de serie temporal
fechas = pd.date_range(start="2023-01-01", periods=50, freq="D")
ventas = np.cumsum(np.random.randn(50) * 10 + 50)  # Ventas acumuladas

# Crear DataFrame
df = pd.DataFrame({"Fecha": fechas, "Ventas": ventas})

# Crear gráfico de líneas
sns.lineplot(x="Fecha", y="Ventas", data=df, color="green")
plt.title("Tendencia de Ventas a lo largo del tiempo")
plt.xticks(rotation=45)
plt.xlabel("Fecha")
plt.ylabel("Ventas")
plt.show()
```

3️⃣ Regplot (Gráfico de Regresión)
📌 Descripción expandida

Un regplot es un scatter plot con una línea de regresión ajustada, lo que ayuda a visualizar la tendencia y relación entre dos variables.
📌 Cuándo Usarlo

✅ Para analizar relaciones lineales entre dos variables.
✅ Para ver la tendencia general de los datos.
✅ Para evaluar la fuerza de la correlación entre variables.
📌 Excepciones

❌ No es ideal para relaciones no lineales.
❌ Puede ser engañoso si hay muchos valores atípicos.
📌 Código Base

```python
sns.regplot(x=variable_x, y=variable_y, data=datos)
```

📌 Explicación del Código:

    x=variable_x: Variable en el eje X.
    y=variable_y: Variable en el eje Y.
    data=datos: Conjunto de datos usado.

📌 Ejemplo

```python
# Generar datos
x = np.random.rand(100) * 10
y = 2 * x + np.random.randn(100) * 2  # Relación lineal con ruido

# Crear DataFrame
df = pd.DataFrame({"X": x, "Y": y})

# Crear gráfico de regresión
sns.regplot(x="X", y="Y", data=df, color="red")
plt.title("Gráfico de Regresión Lineal")
plt.xlabel("Variable X")
plt.ylabel("Variable Y")
plt.show()
```

4️⃣ Jointplot (Gráfico Conjunta de Relación)
📌 Descripción expandida

Un jointplot combina un scatter plot con histogramas marginales para mostrar la relación entre dos variables y sus distribuciones.
📌 Cuándo Usarlo

✅ Para visualizar correlaciones junto con la distribución de las variables.
✅ Para encontrar patrones ocultos en los datos.
✅ Para identificar si los datos siguen una distribución normal.
📌 Excepciones

❌ No es útil para datos categóricos.
❌ Puede no ser claro si los datos están demasiado dispersos.
📌 Código Base

```python
sns.jointplot(x=variable_x, y=variable_y, data=datos, kind="scatter")
```

📌 Explicación del Código:

    x=variable_x: Variable del eje X.
    y=variable_y: Variable del eje Y.
    data=datos: Conjunto de datos.
    kind="scatter": Define el tipo de gráfico (puede ser "scatter", "kde", "hist").

📌 Ejemplo

```python
# Crear jointplot
sns.jointplot(x="ingresos", y="gastos", data=datos, kind="scatter", color="purple")
plt.show()
```

💡 Consejos Finales

    Usa scatter plot si quieres ver correlaciones sin tendencias forzadas.
    Si trabajas con series temporales, usa line plot.
    Para evaluar correlación lineal, regplot es ideal.
    Si necesitas visualizar distribución y correlación juntas, usa jointplot.
