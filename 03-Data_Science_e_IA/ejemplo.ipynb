游늵 Tipos de Gr치ficos en Python: Conceptos, Ejemplos y Cu치ndo Usarlos

游댳 1. Gr치fico de L칤neas (plot)

游댌 Concepto:

Los gr치ficos de l칤neas son ideales para representar datos continuos a lo largo del tiempo, mostrando tendencias y patrones.

游늰 Cu치ndo Usarlo:

Para analizar series temporales.

Para observar la tendencia de una variable a lo largo del tiempo.

Comparar m칰ltiples tendencias en un mismo gr치fico.

游댃 Ejemplo:

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y, label='Seno', color='blue')
plt.title("Gr치fico de L칤neas")
plt.xlabel("Tiempo")
plt.ylabel("Valor")
plt.legend()
plt.show()
```

游댳 2. Gr치fico de Barras (bar)

游댌 Concepto:

Se usa para comparar diferentes categor칤as o mostrar la frecuencia de ocurrencias de diferentes elementos.

游늰 Cu치ndo Usarlo:

Comparar cantidades entre diferentes categor칤as.

Representar datos categ칩ricos.

Mostrar la frecuencia de valores discretos.

游댃 Ejemplo:

```python
categorias = ["A", "B", "C", "D"]
valores = [5, 7, 3, 8]

plt.bar(categorias, valores, color='purple')
plt.title("Gr치fico de Barras")
plt.xlabel("Categor칤as")
plt.ylabel("Valores")
plt.show()
```

游댳 3. Histograma (hist)

游댌 Concepto:

Muestra la distribuci칩n de una variable num칠rica dividi칠ndola en intervalos (bins).

游늰 Cu치ndo Usarlo:

Analizar la distribuci칩n de una variable.

Determinar si una variable sigue una distribuci칩n normal.

游댃 Ejemplo:

```python
datos = np.random.randn(1000)

plt.hist(datos, bins=30, color='green', alpha=0.7)
plt.title("Histograma de Distribuci칩n")
plt.xlabel("Valores")
plt.ylabel("Frecuencia")
plt.show()
```

游댳 4. Gr치fico de Dispersi칩n (scatter)

游댌 Concepto:

Muestra la relaci칩n entre dos variables num칠ricas con puntos dispersos.

游늰 Cu치ndo Usarlo:

游댃 Ejemplo:

```python
x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x, y, color='red', alpha=0.6)
plt.title("Gr치fico de Dispersi칩n")
plt.xlabel("Variable X")
plt.ylabel("Variable Y")
plt.show()
```

游댳 5. Gr치fico de Torta (pie)

游댌 Concepto:

Se usa para representar proporciones o porcentajes en un conjunto de datos.

游늰 Cu치ndo Usarlo:

Representar porcentajes de una variable categ칩rica.

Comparar la participaci칩n de cada categor칤a en el total.

游댃 Ejemplo:

```python
etiquetas = ['A', 'B', 'C', 'D']
valores = [30, 25, 35, 10]

plt.pie(valores, labels=etiquetas, autopct='%1.1f%%', colors=['gold', 'blue', 'red', 'green'])
plt.title("Gr치fico de Torta")
plt.show()
```

2勇 救넖잺 Gr치ficos de Distribuci칩n en Python: Conceptos, Ejemplos y Cu치ndo Usarlos

En esta secci칩n, exploraremos los diferentes tipos de gr치ficos de distribuci칩n utilizados en Python para visualizar la distribuci칩n de datos, identificaci칩n de valores at칤picos y patrones en un conjunto de datos.

游댳 1. Boxplot (Diagrama de Cajas) (boxplot)

游댌 Concepto:

Un boxplot muestra la distribuci칩n de una variable mediante un resumen estad칤stico (m칤nimo, primer cuartil, mediana, tercer cuartil y m치ximo). Tambi칠n permite identificar valores at칤picos.

游늰 Cu치ndo Usarlo:

Para visualizar la distribuci칩n y dispersi칩n de una variable num칠rica.

Para identificar valores at칤picos (outliers).

Para comparar distribuciones entre diferentes grupos.

游댃 Ejemplo:

```python
import matplotlib.pyplot as plt
import numpy as np

datos = np.random.randn(100)

plt.boxplot(datos)
plt.title("Boxplot - Diagrama de Cajas")
plt.ylabel("Valores")
plt.show()
```

游댳 2. Violin Plot (violinplot)

游댌 Concepto:

Un violin plot combina la informaci칩n del boxplot con una distribuci칩n de densidad, proporcionando una visi칩n m치s detallada de la distribuci칩n de los datos.

游늰 Cu치ndo Usarlo:

Cuando se necesita analizar la forma y simetr칤a de la distribuci칩n.

Para comparar distribuciones entre varios grupos.

Para ver si los datos tienen distribuci칩n bimodal o multimodal.

游댃 Ejemplo:

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

datos = np.random.randn(100)
sns.violinplot(y=datos)
plt.title("Violin Plot")
plt.show()
```

游댳 3. Stem Plot (stem)

游댌 Concepto:

Un stem plot representa datos discretos mediante l칤neas verticales desde el eje X, facilitando la visualizaci칩n de distribuciones de valores en ciertos puntos.

游늰 Cu치ndo Usarlo:

Cuando se quiere visualizar series temporales con valores discretos.

Para destacar valores individuales en una serie de datos.

Para analizar patrones en datos con puntos espec칤ficos.

游댃 Ejemplo:

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 10)
y = np.sin(x)

plt.stem(x, y)
plt.title("Stem Plot")
plt.xlabel("Tiempo")
plt.ylabel("Valores")
plt.show()
```

游댳 4. Gr치fico de Pir치mide Poblacional

游댌 Concepto:

Este tipo de gr치fico se utiliza para comparar distribuciones en diferentes categor칤as, generalmente en estudios demogr치ficos.

游늰 Cu치ndo Usarlo:

Para analizar la distribuci칩n de poblaciones por edad y g칠nero.

Para comparar dos grupos en una misma categor칤a.

En estudios de mercado para analizar segmentos de clientes.

游댃 Ejemplo:

```python
import matplotlib.pyplot as plt

categorias = ['0-18', '19-35', '36-50', '51+']
mujeres = [30, 40, 35, 20]
hombres = [35, 45, 30, 25]

plt.barh(categorias, mujeres, color='pink', label='Mujeres')
plt.barh(categorias, [-h for h in hombres], color='blue', label='Hombres')
plt.axvline(0, color='black', linewidth=0.8)
plt.title("Pir치mide Poblacional")
plt.legend()
plt.show()
```

3勇 游 Gr치ficos Avanzados en Python: Conceptos, Ejemplos y Cu치ndo Usarlos

En esta secci칩n, exploraremos los diferentes tipos de gr치ficos avanzados utilizados en Python para visualizar datos de manera m치s detallada y obtener insights profundos sobre patrones y relaciones en los datos.

游댳 1. Heatmap (Mapa de Calor) (imshow)

游댌 Concepto:

Un heatmap es una representaci칩n en color de datos organizados en una matriz, donde los colores indican la magnitud de los valores.

游늰 Cu치ndo Usarlo:

Para analizar correlaciones entre variables en un dataset.

Para visualizar distribuciones de datos en una matriz.

Para analizar datos en cuadr칤culas (ejemplo: modelos geoespaciales o de redes neuronales).

游댃 Ejemplo:

```python
import matplotlib.pyplot as plt
import numpy as np

matriz = np.random.rand(10, 10)
plt.imshow(matriz, cmap='viridis', interpolation='nearest')
plt.colorbar()
plt.title("Heatmap - Mapa de Calor")
plt.show()
```

游댳 2. Gr치fico de Contorno (contour)

游댌 Concepto:

El gr치fico de contorno muestra las l칤neas de nivel de una funci칩n 3D en un espacio 2D.

游늰 Cu치ndo Usarlo:

Para representar superficies en 3D en un plano 2D.

Para visualizar distribuciones de densidad en modelos de datos continuos.

En aplicaciones geoespaciales y modelado cient칤fico.

游댃 Ejemplo:

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

plt.contour(X, Y, Z, levels=10, cmap='plasma')
plt.title("Gr치fico de Contorno")
plt.colorbar()
plt.show()
```

游댳 3. Gr치fico de Contorno Relleno (contourf)

游댌 Concepto:

Similar al gr치fico de contorno, pero con 치reas rellenas de color para representar los niveles de magnitud.

游늰 Cu치ndo Usarlo:

Para resaltar mejor la diferencia entre regiones de valores en un conjunto de datos continuos.

Para visualizar superficies de energ칤a en modelado cient칤fico.

游댃 Ejemplo:

```python
plt.contourf(X, Y, Z, levels=20, cmap='coolwarm')
plt.title("Gr치fico de Contorno Relleno")
plt.colorbar()
plt.show()
```

游댳 4. Quiver Plot (quiver)

游댌 Concepto:

El quiver plot representa un campo vectorial en un espacio bidimensional mediante flechas.

游늰 Cu치ndo Usarlo:

Para representar direcciones y magnitudes en un campo vectorial (ejemplo: vientos en meteorolog칤a).

En modelado de flujos de fluidos y simulaciones f칤sicas.

游댃 Ejemplo:

```python
X, Y = np.meshgrid(np.arange(0, 10, 1), np.arange(0, 10, 1))
U = np.sin(X)
V = np.cos(Y)

plt.quiver(X, Y, U, V)
plt.title("Quiver Plot - Campo Vectorial")
plt.show()
```

游댳 5. Streamplot (streamplot)

游댌 Concepto:

El streamplot muestra el flujo de part칤culas en un campo vectorial continuo.

游늰 Cu치ndo Usarlo:

Para analizar flujos de fluidos en din치mica de fluidos computacional.

Para representar direcciones y trayectorias en simulaciones f칤sicas.

游댃 Ejemplo:

```python
Y, X = np.mgrid[0:10:100j, 0:10:100j]
U = np.cos(X)
V = np.sin(Y)

plt.streamplot(X, Y, U, V, color=U, linewidth=2, cmap='viridis')
plt.title("Streamplot - Flujo de Part칤culas")
plt.show()
```

游댳 6. Error Bars (errorbar)

游댌 Concepto:

El error bar plot agrega barras de error a los puntos de datos para representar la incertidumbre de las mediciones.

游늰 Cu치ndo Usarlo:

En experimentos cient칤ficos donde hay incertidumbre en las mediciones.

Para visualizar intervalos de confianza en un conjunto de datos.

游댃 Ejemplo:

```python
x = np.linspace(0, 10, 10)
y = np.sin(x)
errores = np.random.rand(10) * 0.1

plt.errorbar(x, y, yerr=errores, fmt='o', capsize=5)
plt.title("Error Bars - Barras de Error")
plt.show()
```
